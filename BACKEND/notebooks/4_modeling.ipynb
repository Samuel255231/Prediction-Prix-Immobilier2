# ETAPES 1 : UTILISATION DES CHAQUES MODELES

## **MODEL 1 : Linear Regression**
Concept

Linear regression est model mathématique simple.

* Relation: y= a*x + b
* Avantages :rapide ;interprétable
* Limite : seulement pour la relation linéaire

**Training**

```python
from sklearn.linear_model import LinearRegression   # import model

lr = LinearRegression()   # création modèle

lr.fit(X_train, y_train)  # entrainement modèle

y_pred_lr = lr.predict(X_test)   # prédire prix maisons

y_pred_lr[:10]  # afficher les 5 premières prédictions
```

**Evaluation**

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# R2
r2_lr = r2_score(y_test, y_pred_lr)

# MSE
mse_lr = mean_squared_error(y_test, y_pred_lr)

# RMSE
rmse_lr = np.sqrt(mse_lr)

# MAE
mae_lr = mean_absolute_error(y_test, y_pred_lr)

print("R2:", r2_lr)
print("RMSE:", rmse_lr)
print("MAE:", mae_lr)
```

Random Forest = ensemble arbres décision.

Principe:

## **I. CROSS VALIDATION (K-Fold)**
Principe de la Cross-Validation (K-Fold)
* Diviser les données : On découpe le jeu de données en K parties (folds) égales.
* Entraînement et test :
1. On utilise K-1 folds pour entraîner le modèle.
2. On utilise le fold restant pour tester le modèle.
3. Répéter K fois : Chaque fold sert une fois de test et K-1 fois d'entraînement.

* Calculer la performance : On prend la moyenne des scores obtenus sur les K tests → cela donne une estimation plus fiable de la performance du modèle.

✅ Avantage : Évite le sur-apprentissage sur le jeu de test et donne une évaluation robuste.

```python
from sklearn.model_selection import cross_val_score # fonction cross validation
from sklearn.ensemble import RandomForestRegressor # Model  test

# Creation de model
rf = RandomForestRegressor(
    n_estimators =200 , # Nombres d'arbres
    random_state=42,
    max_depth=None
)

#lancer les  cross validation
cv_scores = cross_val_score(

    rf,        # modèle
    X
    ,         # dataset complet
    y,         # target
    cv=5,      # nombre folds
    scoring="r2"   # metric evaluation
)

#afficher les  résultats
print("Scores folds:", cv_scores)
print("R2 moyen:", cv_scores.mean())
print("Ecart-type:", cv_scores.std())
```

## **II. HYPERPARAMETER TUNING**
 Consiste à tester différentes valeurs des hyperparamètres d’un modèle afin de trouver la

 combinaison qui donne les meilleures performances lors de l’entraînement et de l’évaluation.

| Paramètre         | rôle             |
| ----------------- | ---------------- |
| n_estimators      | nombre arbres    |
| max_depth         | profondeur arbre |
| min_samples_split | condition split  |


Problem:

Quelle valeur choisir ?

 **GRID SEARCH**

* Principe:

tester plusieurs combinaisons paramètres.

n_estimators = 100,200

max_depth = 5,10

## **GridSearch CV**

```python
# code pour la GridsearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

#Definir le model
rf = RandomForestRegressor(random_state=42)

#définir grille paramètres
param_grid = {

    "n_estimators":[100,200,300],

    "max_depth":[5,10,20,None], # profondeur

    "min_samples_split":[2,5,10] # minimum data split
}

#Lancement de GridSearchCV

grid = GridSearchCV(

    rf,             # modèle
    param_grid,     # grille paramètres
    cv=5,           # cross validation
    scoring="r2",   # metric optimisation
    n_jobs=-1       # utiliser CPU
)

#eEntrainer ou train
grid.fit(X_train,y_train)

#Afficher les meileure modeles
print("Best parameters:", grid.best_params_)

print("Best score:", grid.best_score_)
```

## **1. Ensemble Learning (Concept)**
Ensemble learning = combinaison de differents modèles  pour avooir de bonne prediction .

Principe:

model 1 → prediction

model 2 → prediction

model 3 → prediction

résultat final = combinaison predictions

## **2. Voting Regressor**
Voting = moyenne predictions modèles.

Exemple:

RandomForest → 210000

XGBoost → 205000

Ridge → 200000

prediction finale ≈ 205000

```python
from sklearn.ensemble import VotingRegressor
from sklearn.metrics import r2_score,mean_squared_error

#définir modèles
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

ridge = Ridge(alpha=1.0)

rf = RandomForestRegressor( n_estimators=200,random_state=42 )

xgb = XGBRegressor(n_estimators=300, learning_rate=0.05,max_depth=6 )

#créer Voting Regressor
voting_model = VotingRegressor(estimators=[('ridge', ridge), ('rf', rf),('xgb', xgb) ])

#training
voting_model.fit(X_train, y_train)

#Prediction
y_pred_vote = voting_model.predict(X_test)

#Evaluation
r2_vote = r2_score(y_test, y_pred_vote)
#rmse_vote = root_mean_squared_error(y_test, y_pred_vote, squared=False)
rmse_vote = np.sqrt(mean_squared_error(y_test, y_pred_vote) )# RMSE (racine carrée de MSE)
print("r2_vote :",r2_vote)
```

## **3. Bagging Regressor**
Bagging = Bootstrapping + Aggregation.
* Principe:
créer plusieurs sous-datasets, entraîner modèle sur chaque dataset ,moyenne predictions.

Example:

* dataset1 → modèle
* dataset2 → modèle
* dataset3 → modèle

prediction finale = moyenne

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor

# Créer  de modele
bag_model = BaggingRegressor(

    estimator=DecisionTreeRegressor(),#modèle de base

    n_estimators=100,  # nombre modèles

    random_state=42
)

# Training
bag_model.fit(X_train, y_train)

# Prediction
y_pred_bag = bag_model.predict(X_test)

#Evaluation
r2_bag = r2_score(y_test, y_pred_bag) # R²
rmse_bag = np.sqrt(mean_squared_error(y_test, y_pred_bag))# RMSE (racine carrée de MSE)

print(f"R²: {r2_bag:.4f}")
print(f"RMSE: {rmse_bag:.4f}")
```

## **4. Boosting (Gradient / XGBoost)**
Boosting = modèles successifs qui corrigent erreurs.

Process:

model1 → erreurs

model2 corrige erreurs

model3 corrige erreurs restantes

```python
from sklearn.ensemble import  GradientBoostingRegressor

# Creation de modele
gbr = GradientBoostingRegressor(

    n_estimators=150,
    learning_rate=0.05,
    max_depth=3,
    random_state=42
)

#ENtrennement
gbr.fit(X_train, y_train)

# Prediction
y_pred_gbr = gbr.predict(X_test)

# Evaluation
r2_gbr = r2_score(y_test, y_pred_gbr) # R²
rmse_gbr = np.sqrt(mean_squared_error(y_test, y_pred_gbr))# RMSE (racine carrée de MSE)

print(f"R²: {r2_gbr:.4f}")
print(f"RMSE: {rmse_gbr:.4f}")
```

## **5. Stacking Regressor**
Stacking = model meta combine predictions modèles base.
Architecture:

Model 1 → prediction

Model 2 → prediction

Model 3 → prediction

↓
Meta Model

prediction finale

```python
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import LinearRegression

#Création de modele
stack_model = StackingRegressor(estimators=[('rf', rf),('xgb', xgb),('ridge', ridge)],

    final_estimator=LinearRegression()
)

# Training
stack_model.fit(X_train, y_train)

# Prediction
y_pred_stack = stack_model.predict(X_test)

#Evaluation

r2_stack = r2_score(y_test, y_pred_stack) # R²
rmse_stack = np.sqrt(mean_squared_error(y_test, y_pred_stack))# RMSE (racine carrée de MSE)

print(f"R²: {r2_stack:.4f}")
print(f"RMSE: {rmse_stack:.4f}")
```

# **6. Comparaison finale**

```python
results = pd.DataFrame({

"Model":["RandomForest","XGBoost","Voting","Bagging","GradientBoosting","Stacking"],

"R2":[r2_rf,r2_xgb,r2_vote,r2_bag,r2_gbr,r2_stack],

"RMSE":[rmse_rf,rmse_xgb,rmse_vote,rmse_bag,rmse_gbr,rmse_stack ]

})

results.sort_values("R2", ascending=False)
```
